```python
import pandas as pd
```

```python
men_2024 = pd.read_table('results_men_2024.txt', sep=',')
men_2024.head()
```

```python
men_2019 = pd.read_table('results_men_2019.txt', sep=',')
men_2019.head()
```

```python
# Source: http://registration.baa.org/2014/cf/Media/iframe_ResultsSearch.cfm?mode=advanceddetail
men_2014 = pd.read_table('results_men_2014.txt', sep=',')
men_2014.head()
```

```python
women_2024 = pd.read_table('results_women_2024.txt', sep=',')
women_2024.head()
```

```python
women_2019 = pd.read_table('results_women_2019.txt', sep=',')
women_2019.head()
```

```python
women_2014 = pd.read_table('results_women_2014.txt', sep=',')
women_2014.head()
```

```python
total_2024 = pd.concat([men_2024, women_2024])
total_2024.shape
```

```python
total_2019 = pd.concat([men_2019, women_2019])
total_2019.shape
```

```python
total_2014 = pd.concat([men_2014, women_2014])
total_2014.shape
```

```python
age_2024 = total_2024.groupby(['AgeOnRaceDay','Gender']).aggregate({'FullName':'count'}).reset_index()
age_2024[age_2024['AgeOnRaceDay'] < 30].groupby('Gender').aggregate({'FullName':'sum'})/total_2024.shape[0]
# age_2024[age_2024['AgeOnRaceDay'] < 30].sum()['FullName']/total_2024.shape[0]
```

```python
age_2019 = total_2019.groupby(['AgeOnRaceDay','Gender']).aggregate({'FullName':'count'}).reset_index()
age_2019[age_2019['AgeOnRaceDay'] < 30].groupby('Gender').aggregate({'FullName':'sum'})/total_2019.shape[0]
# age_2019[age_2019['AgeOnRaceDay'] < 30].sum()['FullName']/total_2019.shape[0]
```

```python
age_2014 = total_2014.groupby(['AgeOnRaceDay','Gender']).aggregate({'FullName':'count'}).reset_index()
age_2014[age_2014['AgeOnRaceDay'] < 30].groupby('Gender').aggregate({'FullName':'sum'})/total_2014.shape[0]
# age_2014[age_2014['AgeOnRaceDay'] < 30].sum()['FullName']/total_2014.shape[0]
```
