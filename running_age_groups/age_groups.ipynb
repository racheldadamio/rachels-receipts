### Columbus marathon

We are looking at the age group distribution for the Columbus marathon over the past 10 years.

```python
res_2023 = {
    'total':3018,
    'men_2024':261,
    'women_2024':176,
    'men_2529':349,
    'women_2529':239
}

res_2022 = {
    'total':2397,
    'men_2024':155,
    'women_2024':102,
    'men_2529':261,
    'women_2529':181
}

res_2021 = {
    'total':1740,
    'men_2024':106,
    'women_2024':74,
    'men_2529':176,
    'women_2529':130
}

res_2019 = {
    'total':3592,
    'men_2024':208,
    'women_2024':180,
    'men_2529':314,
    'women_2529':304
}

res_2018 = {
    'total':3215,
    'men_2024':205,
    'women_2024':165,
    'men_2529':267,
    'women_2529':256
}

res_2017 = {
    'total':3203,
    'men_2024':170,
    'women_2024':191,
    'men_2529':253,
    'women_2529':256
}

res_2016 = {
    'total':3850,
    'men_2024':191,
    'women_2024':221,
    'men_2529':263,
    'women_2529':293
}

res_2015 = {
    'total':4443,
    'men_2024':239,
    'women_2024':260,
    'men_2529':328,
    'women_2529':363
}

res_2014 = {
    'total':5461,
    'men_2024':290,
    'women_2024':325,
    'men_2529':421,
    'women_2529':486
}

res_2013 = {
    'total':5523,
    'men_2024':311,
    'women_2024':343,
    'men_2529':432,
    'women_2529':505
}
```

```python
# Convert dictionaries to a DataFrame and add a year column
df = pd.DataFrame([res_2023, res_2022, res_2021, res_2019, res_2018, res_2017, res_2016, res_2015, res_2014, res_2013], index=[2023, 2022, 2021, 2019, 2018, 2017, 2016, 2015, 2014, 2013]).reset_index()

# Rename the index column to 'year'
df = df.rename(columns={'index': 'year'})

# Display the DataFrame
df
```

```python
# Create columns that add the men's and women's counts for the same group
df['total_2024'] = df['men_2024'] + df['women_2024']
df['total_2529'] = df['men_2529'] + df['women_2529']

# Create a column for the combined total of the 2024 and 2529 groups
df['total_2024_2529'] = df['total_2024'] + df['total_2529']
df['women_2024_2529'] = df['women_2024'] + df['women_2529']
df['men_2024_2529'] = df['men_2024'] + df['men_2529']

# Calculate the percent of total for each count
df['men_2024_percent'] = (df['men_2024'] / df['total']) * 100
df['women_2024_percent'] = (df['women_2024'] / df['total']) * 100
df['men_2529_percent'] = (df['men_2529'] / df['total']) * 100
df['women_2529_percent'] = (df['women_2529'] / df['total']) * 100
df['total_2024_percent'] = (df['total_2024'] / df['total']) * 100
df['total_2529_percent'] = (df['total_2529'] / df['total']) * 100
df['women_2024_2529_percent'] = (df['women_2024_2529'] / df['total']) * 100
df['men_2024_2529_percent'] = (df['men_2024_2529'] / df['total']) * 100
df['total_2024_2529_percent'] = (df['total_2024_2529'] / df['total']) * 100

df
```

```python
# Plotting the percentages with the x variable "year"
plt.figure(figsize=(12, 8))

plt.plot(df['year'], df['men_2024_percent'], marker='o', label='Men 20-24')
plt.plot(df['year'], df['women_2024_percent'], marker='o', label='Women 20-24')
plt.plot(df['year'], df['men_2529_percent'], marker='o', label='Men 25-29')
plt.plot(df['year'], df['women_2529_percent'], marker='o', label='Women 25-29')
plt.plot(df['year'], df['total_2024_percent'], marker='o', label='Total 20-24')
plt.plot(df['year'], df['total_2529_percent'], marker='o', label='Total 25-29')
plt.plot(df['year'], df['men_2024_2529_percent'], marker='o', label='Women 20-29')
plt.plot(df['year'], df['women_2024_2529_percent'], marker='o', label='Men 20-29')
plt.plot(df['year'], df['total_2024_2529_percent'], marker='o', label='Total 20-29')

# Adding data labels
for line in plt.gca().lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        plt.annotate(f'{y:.1f}%', xy=(x, y), textcoords="offset points", xytext=(0, 5), ha='center')

plt.xlabel('Year')
plt.ylabel('Percentage of Total Marathon Participants')
plt.title('Percentage of Total Columbus Marathon Participants in Their 20s Over the Past 10 Years')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Create a figure with 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(16, 6), sharey=True)
#axs.set_title('Percentage of Men Columbus Marathon Participants in Their 20s Over the Past 10 Years')

# Plot for Men
axs[0].plot(df['year'], df['men_2024_percent'], marker='o', color = 'tab:orange', label='20-24')
axs[0].plot(df['year'], df['men_2529_percent'], marker='o', color = 'tab:purple', label='25-29')
axs[0].plot(df['year'], df['men_2024_2529_percent'], marker='o', color = 'tab:grey', label='20-29')

for line in axs[0].lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        axs[0].annotate(f'{y:.1f}%', xy=(x, y), textcoords="offset points", xytext=(0, 5), ha='center')

axs[0].set_title('Men')
axs[0].set_ylabel('Percentage')
axs[0].legend(loc='upper left')
axs[0].grid(True)

# Plot for Women
axs[1].plot(df['year'], df['women_2024_percent'], marker='o', color = 'tab:orange', label='20-24')
axs[1].plot(df['year'], df['women_2529_percent'], marker='o', color = 'tab:purple', label='25-29')
axs[1].plot(df['year'], df['women_2024_2529_percent'], marker='o', color = 'tab:grey', label='20-29')

for line in axs[1].lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        axs[1].annotate(f'{y:.1f}%', xy=(x, y), textcoords="offset points", xytext=(0, 5), ha='center')

axs[1].set_title('Women')
axs[1].grid(True)

# Plot for Total
axs[2].plot(df['year'], df['total_2024_percent'], marker='o', color = 'tab:orange', label='20-24')
axs[2].plot(df['year'], df['total_2529_percent'], marker='o', color = 'tab:purple', label='25-29')
axs[2].plot(df['year'], df['total_2024_2529_percent'], marker='o', color = 'tab:grey', label='20-29')

for line in axs[2].lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        axs[2].annotate(f'{y:.1f}%', xy=(x, y), textcoords="offset points", xytext=(0, 5), ha='center')

axs[2].set_title('Total')
axs[2].grid(True)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.97])  # Adjust rect to make space for suptitle
plt.suptitle('Percentage of Columbus Marathon Participants in Their 20s Over the Past Decade - By Gender', fontsize=14)

# Show the plots
plt.show()
```
