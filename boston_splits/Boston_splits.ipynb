```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
boston_splits = pd.read_csv("Boston_splits.csv", index_col="Distance")
```

```python
boston_corr = boston_splits.corr()[["Karen", "Rachel"]]
```

```python
boston_corr.sort_values("Karen").tail()
```

```python
boston_corr.sort_values("Rachel").tail()
```

```python
boston_splits[["Karen", "Helen Obiri", "Rachel", "Sara Hall"]]
```

```python
distance = boston_splits.index
karen = boston_splits["Karen"]
rachel = boston_splits["Rachel"]
helen = boston_splits["Helen Obiri"]
sara = boston_splits["Sara Hall"]

plt.plot(distance, karen, label="Karen", color="blue", linestyle="-")
plt.plot(distance, helen, label="Helen Obiri", color="green", linestyle="-")
plt.plot(distance, rachel, label="Rachel", color="red", linestyle="--")
plt.plot(distance, sara, label="Sara Hall", color="purple", linestyle="--")

plt.xlabel("Distance")
plt.ylabel("MPH")
plt.title("Boston Marathon 2025 Race Progression")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

plt.plot(distance, karen, label="Karen", color="dodgerblue", linewidth=2)
plt.plot(distance, helen, label="Helen Obiri", color="seagreen", linewidth=2)
plt.plot(distance, rachel, label="Rachel", color="crimson", linestyle="--", linewidth=2)
plt.plot(
    distance, sara, label="Sara Hall", color="mediumpurple", linestyle="--", linewidth=2
)

plt.xlabel("Distance (miles)", fontsize=12)
plt.ylabel("Speed (MPH)", fontsize=12)
plt.title(
    "Boston Marathon 2025: Race Progression by MPH", fontsize=14, fontweight="bold"
)

plt.legend(title="Runners", fontsize=10, title_fontsize=11)

plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```
